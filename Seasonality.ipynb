```python
import xgboost as xgb
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
import joblib
```

1. Spring

```python
data = pd.read_csv('C:/Users/Admin/Desktop/ML/rail data/2019_spring.csv')
data1 = pd.read_csv('C:/Users/Admin/Desktop/ML/rail data/2019_spring_90.csv')

df=data
df1=data1

X_train = df.iloc[:,1:10]
y_train = df.iloc[:,-1]
X_test = df1.iloc[:,1:10]
y_test = df1.iloc[:,-1]

model_spring = xgb.XGBRegressor()
model_spring.fit(X_train,y_train)
predictions = model_spring.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)

print(model_spring.score(X_train, y_train))
print(model_spring.score(X_test, y_test))
print(mae)
print(mse)

xgb.plot_importance(model_spring)

joblib.dump(model_spring,'model_spring.pkl')
```

Ensemble

```python
from sklearn.ensemble import VotingRegressor, RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
import numpy as np

data = pd.read_csv('C:/Users/Admin/Desktop/ML/rail data/real_weather.csv')
df = data

X = df.iloc[:,1:10]
y = df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
estimator=[
('model_spring', model_spring),
('model_summer', model_summer),
('model_fall', model_fall),
('model_winter', model_winter)
]
```

1. Voting

```python
voting= VotingRegressor(estimators=estimator)
voting.fit(X_train, y_train)
```

```python
predictions = voting.predict(X_test)

score = voting.score(X_test, y_test)
mse = metrics.mean_squared_error(predictions, y_test)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(predictions, y_test)
print(score)
print(rmse)
print(r2)
```

```python
joblib.dump(voting,'voting.pkl')
```
